```python
import import_ipynb
import wikiTQ_HowtoUse

from datasets import load_dataset
import time
import pandas as pd
import re
import os
import json
import requests
import torch
import numpy as np

# 데이터셋 로드
dataset = load_dataset("Stanford/wikitablequestions", trust_remote_code=True)

# 데이터셋 구조 확인
# print(dataset)
```

```python
from llama_index.llms.openai import OpenAI
import os

os.environ["OPENAI_API_KEY"] = "OPENAI_API_KEY"

# 모델 이름을 명시하지 않으면, query때 자동으로 gpt-3.5 turbo를 씀
llm = OpenAI(model="gpt-4o-mini-2024-07-18")
```

```python
"""
Tbh chain-of-table is a cool paper and concept, but not overly reliable unless you use GPT-4    
"""

index_list = [i for i in range(1, 11)]
df_list = []
question_list = []
answer_list = []
# 객체 생성
missed_count = 0
correct_count = 0
for index in index_list:
    wiki_tq = wikiTQ_HowtoUse.WikiTQ(dataset, index)

    # 메서드 사용 예시
    df = wiki_tq.get_table()
    question = wiki_tq.get_question()
    answer = wiki_tq.get_answers()
    
    print(question)
    
    try:
    
        query_engine = ChainOfTableQueryEngine(df, llm=llm)
        response = query_engine.query(question)
        
        print(f'The query engine answered: {str(response)}\nThe answer is {answer}\n')
        
        correct_count += 1
    except:
        print(f'LLM did not answer the correctly.\n')
        missed_count += 1

print(f'output_count: {correct_count}')
print(f'missed_count: {missed_count}')
    
```
